Se divide el conjunto de datos en train y test para que pueda ser usado por los distintos algoritmos.

Se deja para test los datos con los últimos 3 meses.

```python
import math
import itertools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from dateutil.relativedelta import relativedelta
from datetime import datetime
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/Trabajo profesional/Code/data/prod_norm_2024-05-27_03:42:59.csv", index_col=0, low_memory=False)
dataset["fecha"] = pd.to_datetime(dataset["fecha"])
dataset.sort_values(by=['fecha', 'idpozo'], ascending=True, inplace=True)
dataset.reset_index(drop=True, inplace=True)
dataset.info()
```

```python
max_date = dataset["fecha"].max() # fecha maxima de todo el dataset
cutoff_date = max_date + relativedelta(months=-3)

print(f"Se deja el intervalo de datos entre {cutoff_date.strftime('%d/%m/%Y')} y {max_date.strftime('%d/%m/%Y')} como test (últimos 3 meses)")

mask = dataset["fecha"] > cutoff_date

train = dataset[~mask]
test = dataset[mask]

print(f"El dataset para entrenamiento queda con {len(train)} registros")
print(f"El dataset para pruebas queda con {len(test)} registros")
```

# Export

```python
datetime_now = datetime.now().strftime("_%Y-%m-%d_%H:%M:%S")

# Estaba en duda de reiniciar los indices, pero se los dejo por si en otra etapa se quiere volver a unir los datasets

train.sort_values(by=['fecha', 'idpozo'], ascending=True, inplace=True)
train.to_csv("/content/drive/MyDrive/Trabajo profesional/Code/data/prod_split_train" + datetime_now + ".csv")

test.sort_values(by=['fecha', 'idpozo'], ascending=True, inplace=True)
test.to_csv("/content/drive/MyDrive/Trabajo profesional/Code/data/prod_split_test" + datetime_now + ".csv")
```
