```python
#ME QUEDO CON LOS DATOS ENTRE 2020 Y 2023

between2020And2023 = prod_data.loc[(prod_data['anio'] > 2019) & (prod_data['anio'] < 2025) & (prod_data['tipopozo'] == 'Petrolífero')]
```

```python
#ME QUEDO CON LOS POZOS PETROLIFEROS QUE TENGAN 51 MESES DE DATOS

conteo_pozos = between2020And2023['idpozo'].value_counts()
data_pozos = between2020And2023[between2020And2023['idpozo'].isin(conteo_pozos[conteo_pozos == 51].index)].sort_values(by=['idpozo','anio'])
data_pozos
```

```python
POZOS = data_pozos['idpozo'].unique()
WIDTH = 51
COPIES = 40
```

```python
data = data_pozos['prod_pet'].values.reshape(len(POZOS), 51)[:,-1*WIDTH:]
data.shape
```

```python
x_data = np.zeros((len(POZOS)*COPIES, 9)) # X_TRAIN TIME SERIES ARE 9 MONTHS
y_data = np.zeros((len(POZOS)*COPIES, 3)) # Y_TRAIN TIME SERIES ARE 3 MONTHS
x_data.shape, y_data.shape
```

```python
for j in range(len(POZOS)):
    for k in range(COPIES):
        i = j*COPIES + k
        x_data[i,] = data[j,k:k+(9)]
        y_data[i,] = data[j,k+(9):k+(WIDTH-COPIES+1)]
```

```python
# STANDARIZE
mn = np.mean(x_data)
sd = np.std(x_data)
x_data = (x_data - mn)/sd
y_data = (y_data - mn)/sd
```

```python
FEATURES = [f'f{x}' for x in range(9)]
TARGETS = [f'y{x}' for x in range(3)]
train_data = pd.DataFrame(x_data,columns=FEATURES)
train_data[TARGETS] = y_data
train_data['idpozo'] = np.repeat(POZOS,COPIES)
print('Our GRU training data has shape:', train_data.shape )
train_data.head()
```

```python
def build_model():

    inp = tf.keras.Input(shape=(9,1)) # INPUT SHAPE IS 9

    x = tf.keras.layers.GRU(units=32, return_sequences=False, dropout=0.1)(inp)
    x = tf.keras.layers.Dense(3,activation='linear')(x) # OUTPUT SHAPE IS 3
    model = tf.keras.Model(inputs=inp, outputs=x)

    opt = tf.keras.optimizers.Adam(learning_rate=1e-4)
    loss = tf.keras.losses.MeanSquaredError()
    model.compile(loss=loss, optimizer = opt)

    return model
```

```python
model = build_model()
model.summary()
```

```python
FOLDS = 20
VERBOSE = 2
oof = np.zeros((len(train_data),3))

skf = GroupKFold(n_splits=FOLDS)
skf = skf.split(train_data, train_data['y0'], train_data['idpozo'])
for fold,(train_idx, valid_idx) in enumerate(skf):

    print('#'*25)
    print('### Fold',fold+1)
    print('### Train size',len(train_idx),'Valid size',len(valid_idx))
    print('#'*25)

    # TRAIN, VALID, HOLDOUT FOR FOLD K
    X_train = train_data.loc[train_idx, FEATURES]
    y_train = train_data.loc[train_idx, TARGETS]
    X_valid = train_data.loc[valid_idx, FEATURES]
    y_valid = train_data.loc[valid_idx, TARGETS]

    # WEIGHT RECENT SAMPLE MORE
    GRP = len(train_idx)//COPIES
    w = np.array( [1] * (COPIES-7) + [1,1,1,1] + [2,2,2] )
    w = w/np.sum(w)

    model = build_model()

    if INFER_FROM_PATH is None:
        h = model.fit(X_train,y_train,
                      validation_data = (X_valid,y_valid),
                      sample_weight = np.tile(w,GRP),
                      batch_size=8, epochs=4, verbose=VERBOSE)
    else:
        model.load_weights(INFER_FROM_PATH + f'GRU2_f{fold}_v-2024.h5')
    model.save_weights(f'/content/drive/MyDrive/Trabajo profesional/Code/data/models/GRU2_f{fold}_v-2024.h5')

    oof[valid_idx,] = model.predict(X_valid, verbose=VERBOSE)
```

```python
def smape(y_true, y_pred):

    # CONVERT TO NUMPY
    y_true = np.array(y_true.flatten())
    y_pred = np.array(y_pred.flatten())

    # WHEN BOTH EQUAL ZERO, METRIC IS ZERO
    both = np.abs(y_true) + np.abs(y_pred)
    idx = np.where(both==0)[0]
    y_true[idx]=1; y_pred[idx]=1

    return 100/len(y_true) * np.sum(2 * np.abs(y_pred - y_true) / (np.abs(y_true) + np.abs(y_pred)))
```
