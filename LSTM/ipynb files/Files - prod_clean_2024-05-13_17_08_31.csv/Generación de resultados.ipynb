```python
import math
import itertools
from datetime import datetime
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)

!pip install chardet

import chardet

from google.colab import drive
drive.mount('/content/drive', force_remount=True)

file_path = '/content/drive/MyDrive/Trabajo profesional/Code/LSTM/Predicciones/LSTM_results.csv'

#dataset = pd.read_csv(file_path)

# Detect encoding
with open(file_path, 'rb') as f:
    raw_data = f.read()
    result = chardet.detect(raw_data)
    encoding = result['encoding']
    print(f"Detected encoding: {encoding}")

# Read the CSV file using the detected encoding
dataset = pd.read_csv(file_path, encoding=encoding, sep=';', header=0)
#dataset = pd.read_csv(file_path, encoding=encoding, header=0)
print("Original DataFrame:")
print(dataset.head())

# Drop columa 'Unnamed: 4'
if 'Unnamed: 4' in dataset.columns:
    dataset = dataset.drop(columns=['Unnamed: 4'])

# Drop todas las columnas que empiecen con 'Unnamed:'
dataset = dataset.loc[:, ~dataset.columns.str.startswith('Unnamed:')]

# Reemplazar las comas por puntos en la columna 'prod_pet'
dataset['prod_pet'] = dataset['prod_pet'].str.replace(',', '.')

# Convertir los valores de 'prod_pet' a floats
dataset['prod_pet'] = dataset['prod_pet'].astype(float)

print("Feature names (Column names):")
print(dataset.columns)

column_names = dataset.columns.tolist()
print("Feature names as list:")
print(column_names)

dataset.info()
dataset.head()
```

```python
column_names = ['fecha', 'idpozo', 'prod_pet', 'Modelo']
#dataset = pd.read_csv(file_path, encoding=encoding, names=column_names)

# Seleccionar características relevantes
features = ['fecha', 'idpozo', 'prod_pet', 'Modelo']

# Filtrar características
data = dataset[features]

# Modificar la columna 'idpozo' removiendo el sufijo '.0'
data['idpozo'] = data['idpozo'].astype(str).str.rstrip('.0')
data['idpozo'] = data['idpozo'].replace('1622', '162200')
data['idpozo'] = data['idpozo'].replace('16081', '160810')

filtered_data_MSE_50_trials = data[data['Modelo'] == 'LSTM MSE 50 TRIALS']
# Agregar nueva columna para enumerar los valores: 0, 1, 2, ... a la izquierda del DF
filtered_data_MSE_50_trials.insert(0, '', range(len(filtered_data_MSE_50_trials)))

filtered_data_MSE_MAE_100_trials = data[data['Modelo'] == 'LSTM MSE y MAE 100 TRIALS']
# Agregar nueva columna para enumerar los valores: 0, 1, 2, ... a la izquierda del DF
filtered_data_MSE_MAE_100_trials.insert(0, '', range(len(filtered_data_MSE_MAE_100_trials)))

filtered_data_MSE_100_trials = data[data['Modelo'] == 'LSTM MSE 100 TRIALS']
# Agregar nueva columna para enumerar los valores: 0, 1, 2, ... a la izquierda del DF
filtered_data_MSE_100_trials.insert(0, '', range(len(filtered_data_MSE_100_trials)))

filtered_data_MSE_50_trials.drop(['Modelo'], axis=1, inplace=True)
filtered_data_MSE_MAE_100_trials.drop(['Modelo'], axis=1, inplace=True)
filtered_data_MSE_100_trials.drop(['Modelo'], axis=1, inplace=True)

file_path = '/content/drive/MyDrive/Trabajo profesional/Code/LSTM/Predicciones/predictions_MSE_50_trials.csv'
filtered_data_MSE_50_trials.to_csv(file_path, index=False)

file_path = '/content/drive/MyDrive/Trabajo profesional/Code/LSTM/Predicciones/predictions_MSE_MAE_100_trials.csv'
filtered_data_MSE_MAE_100_trials.to_csv(file_path, index=False)

file_path = '/content/drive/MyDrive/Trabajo profesional/Code/LSTM/Predicciones/predictions_MSE_100_trials.csv'
filtered_data_MSE_100_trials.to_csv(file_path, index=False)
```
