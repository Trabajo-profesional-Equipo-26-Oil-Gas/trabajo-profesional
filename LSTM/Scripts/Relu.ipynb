```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate input data (x values)
x = np.linspace(-10, 10, 100)  # Generate 100 points from -10 to 10

# Step 2: Calculate output data (ReLU function)
relu_values = np.maximum(0, x)  # Applying ReLU function to each element of x

# Step 3: Create a pandas DataFrame for plotting
df = pd.DataFrame({'x': x, 'ReLU(x)': relu_values})

# Step 4: Plot using pandas plot function
df.plot(x='x', y='ReLU(x)', figsize=(8, 6))
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.show()
```
