```python
df = pd.read_csv("/content/drive/MyDrive/Trabajo profesional/Code/data/prod_merge_2024-03-30_02:38:49.csv", index_col=0)
```

```python
# Calculate the number of rows to delete (approximately 20%)
num_rows_to_delete = int(0.2 * len(df))

# Randomly select rows to delete
rows_to_delete_indices = random.sample(range(len(df)), num_rows_to_delete)

# Save the rows to be deleted to a separate CSV file
deleted_rows_df = df.iloc[rows_to_delete_indices]
deleted_rows_file_path = '/content/drive/MyDrive/Trabajo profesional/Code/data/deleted_rows_test_dataset.csv'
deleted_rows_df.to_csv(deleted_rows_file_path, index=False)

# Drop the selected rows from the DataFrame
df = df.drop(rows_to_delete_indices)

# Write the modified DataFrame back to the original CSV file
output_file_path = '/content/drive/MyDrive/Trabajo profesional/Code/data/output_file_train_dataset.csv'
df.to_csv(output_file_path, index=False)

print(f"Deleted {num_rows_to_delete} rows. Modified CSV saved to {output_file_path}.")
print(f"Deleted rows saved to {deleted_rows_file_path}.")
```
