```python
IRRELEVANT_COLUMNS = ['sigla', 'idareapermisoconcesion', 'proyecto', 'idareayacimiento', 'unique_sigla_formprod', 'codigopropio',
                      'nombrepropio', 'observaciones', 'rectificado', 'habilitado', 'idusuario', 'idcuenca', 'idprovincia', 'fecha_data',
                      'fechaingreso', 'tipo_de_recurso', 'tipo_reservorio', 'adjiv_tipo_reservorio', 'tipo_recurso', 'fechadeingreso',
                      'cod_yacimiento', 'cod_area', 'periodo', 'vida_util_inicial', 'vida_util', 'adjiv_equipo_utilizar', 'gasplus']

# Todos los campos siguientes representan valor acumulado hasta el 2005, y este dataset recién incluye pozos desde 2006.
# Además, tomaremos sólo los últimos 10 años, para los cuales todos estos campos tienen valor cero.
IRRELEVANT_COLUMNS.extend(["pet_inicial", "gas_inicial", "agua_inicial",
                           "iny_agua_inicial", "iny_gas_inicial", "iny_otros_inicial", "iny_co2_inicial"])

# Es la sumatoria total de producción para cada pozo, y se repite con igual valor en todos los registros de ese pozo.
# Campos que se pueden derivar de los datos de producción, y en muchos casos son incorrectos pues no coinciden con la sumatoria real.
IRRELEVANT_COLUMNS.extend(["petroleo", "gas", "agua"])

# Con .describe() se constató datos en cero. El percentil p99 es 0 para todos. Promedio es ~cero, dado un outlier máximo.
IRRELEVANT_COLUMNS.extend(["iny_agua", "iny_gas", "iny_co2"])

# Al ser un dato genérico sin especificar, se quita
IRRELEVANT_COLUMNS.extend(["iny_otro"])
```

```python
# Load data from a CSV file into a pandas DataFrame
# Assumes the file is located at "/content/drive/MyDrive/Trabajo profesional/Code/data/sources/listado-de-pozos-cargados-por-empresas-operadoras.csv"
wells_operator_data = pd.read_csv("/content/drive/MyDrive/Trabajo profesional/Code/data/sources/listado-de-pozos-cargados-por-empresas-operadoras.csv")

# Filter the DataFrame to include only wells with "tipo_reservorio" value of "NO CONVENCIONAL"
wells_operator_data = wells_operator_data.query('tipo_reservorio == "NO CONVENCIONAL"')

# List of column names to be removed from the DataFrame
remove_cols = ['areapermisoconcesion', 'areayacimiento', 'idempresa', 'cuenca', 'provincia', 'profundidad',
              'formprod', 'coordenadax', 'coordenaday', 'adjiv_subtipo_reservorio', 'comp_perf', 'clasificacion', 'subclasificacion']

# Remove unwanted columns from the DataFrame using the defined functions
drop_columns(wells_operator_data, IRRELEVANT_COLUMNS)
drop_columns(wells_operator_data, remove_cols)

# Display information about the DataFrame, such as column names, data types, and memory usage
wells_operator_data.info()
```

```python
def replace_nan_with_none(df, inplace):
    """
    Replace NaN (Not a Number) values in a DataFrame with None.

    Parameters:
        df (pandas.DataFrame): The DataFrame in which NaN values are to be replaced.
        inplace: If True, performs operation inplace and returns None.

    Returns:
        df (pandas.DataFrame): If an inplace operation retusn None, else returns a copy of the original DataFrame with NaN values replaced by None.
    """
    return df.replace(math.nan, None, inplace=inplace)
```

```python
def rename_cols_names_suffix_x(df, cols):
    """
    Rename columns in a DataFrame by removing the '_x' suffix.

    Parameters:
        df (pandas.DataFrame): The DataFrame containing columns to be renamed.
        cols (list): A list with the names of the columns to rename without the '_x' suffix.

    Returns:
        None

    """
    cols_rename_map = {}

    for col in cols:
      cols_rename_map[col + "_x"] = col

    df.rename(columns=cols_rename_map, inplace=True)
```

```python
def set_priority_value(df, row, i, col):
    """
    Set a priority value in a DataFrame based on the values from two columns.

    Parameters:
        df (pandas.DataFrame): The DataFrame in which the priority value is to be set.
        row (pandas.Series): The row from which values are to be extracted.
        i (int): The index of the row in the DataFrame.
        col (str): The name of the column without the '_x' suffix.

    Returns:
        None

    """
    # Set the value in the '_x' column to the value from '_x' column if it's not None,
    # otherwise set it to the value from '_y' column
    df.at[i, col + "_x"] = row[col + "_x"] if row[col + "_x"] is not None else row[col + "_y"]
```

```python
# Merge two pandas DataFrames based on a common column 'idpozo'
# The 'how' parameter specifies the type of merge to be performed.
#   - 'outer': Use the union of keys from both DataFrames, similar to a SQL full outer join.
# The 'on' parameter specifies the column to join on.
# The resulting DataFrame will contain all rows from both DataFrames,
# with matching rows combined into a single row and non-matching rows filled with NaN values.
wells_merge_data = wells_iv_data.merge(wells_operator_data, how='outer', on='idpozo')
```

```python
# List of columns to replace NaN values with None
replace_nan_columns = ['adjiv_fecha_inicio_term_x', 'adjiv_fecha_fin_term_x',
                       'adjiv_fecha_inicio_term_y', 'adjiv_fecha_fin_term_y',
                       'adjiv_fecha_inicio_perf', 'adjiv_fecha_fin_perf',
                       'adjiv_fecha_inicio', 'adjiv_fecha_fin']

wells_merge_data[replace_nan_columns] = replace_nan_with_none(wells_merge_data[replace_nan_columns], inplace=False)

# Replace NaN values with None for the specified columns. None does not work with fillna.
# wells_merge[columns_to_replace] = wells_merge[columns_to_replace].fillna(None)
```

```python
for i, row in wells_merge_data.iterrows():
  wells_merge_data.at[i, 'adjiv_fecha_inicio_term_x'] = row['adjiv_fecha_inicio_term_x'] if row['adjiv_fecha_inicio_term_x'] != None else row['adjiv_fecha_inicio_term_y']
  wells_merge_data.at[i, 'adjiv_fecha_fin_term_x'] = row['adjiv_fecha_fin_term_x'] if row['adjiv_fecha_fin_term_x'] != None else row['adjiv_fecha_fin_term_y']
  wells_merge_data.at[i, 'adjiv_fecha_inicio_perf'] = row['adjiv_fecha_inicio_perf'] if row['adjiv_fecha_inicio_perf'] != None else row['adjiv_fecha_inicio']
  wells_merge_data.at[i, 'adjiv_fecha_fin_perf'] = row['adjiv_fecha_fin_perf'] if row['adjiv_fecha_fin_perf'] != None else row['adjiv_fecha_fin']

remove_cols = ['cota_y', 'adjiv_fecha_inicio_term_y', 'adjiv_fecha_fin_term_y', 'adjiv_clasificacion',
               'adjiv_subclasificacion', 'subtipo_reservorio', 'adjiv_fecha_inicio', 'adjiv_fecha_fin']
drop_columns(wells_merge_data, remove_cols)

rename_cols = ["cota", "adjiv_fecha_inicio_term", "adjiv_fecha_fin_term"]
rename_cols_names_suffix_x(wells_merge_data, rename_cols)
```

```python
prod_merge = wells_prod.merge(wells_merge_data, how='outer', on='idpozo')
```

```python
duplicate_cols = ['tipoextraccion','tipoestado','tipopozo','clasificacion','subclasificacion','sub_tipo_recurso']

for col in duplicate_cols:
  replace_nan_with_none(prod_merge[col + '_x'], inplace=True)
  replace_nan_with_none(prod_merge[col + '_y'], inplace=True)

for i, row in prod_merge.iterrows():
  for col in duplicate_cols:
    set_priority_value(prod_merge, row, i, col)

for col in duplicate_cols:
  drop_columns(prod_merge, [col + '_y'])

rename_cols_names_suffix_x(prod_merge, duplicate_cols)
```
